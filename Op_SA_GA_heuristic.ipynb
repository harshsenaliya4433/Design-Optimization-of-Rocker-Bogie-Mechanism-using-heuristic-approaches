```python
import numpy as np
def history_gen(arr):
  traff_hist = []
  arrlen = len(arr)
  for i in range(arrlen):
    traff_hist.append(ctraff(arr[i][4], arr[i][4]))
  p_hist = []
  for i in range(arrlen):
    p_hist.append(P(arr[i][9]))
  clat_hist = []
  alpha = 0 #test
  for i in range(arrlen):
    clat_hist.append(clat(arr[i][1], arr[i][5], arr[i][6], arr[i][7], arr[i][9], alpha)) #Yr, Yb1, Yb2, Zr, c, alpha
  clong_hist = []
  for i in range(arrlen):
    clong_hist.append(clong(arr[i][0], arr[i][2], arr[i][7], alpha)) #Xr, Zr, c, alpha
  e12_hist = []
  for i in range(arrlen):
    e12_hist.append(e1pluse2(arr[i][4], arr[i][4], arr[i][0],  arr[i][8],  arr[i][3])) #Xb2, Xb1, Xr, Lrb, gamma
  thr_hist = []
  for i in range(arrlen):
    thr_hist.append(thetarover(arr[i][3])) #gamma
  return np.asarray(traff_hist), np.asarray(p_hist), np.asarray(clat_hist), np.asarray(clong_hist), np.asarray(e12_hist), np.asarray(thr_hist)
```
